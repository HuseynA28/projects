### Understanding Sample and Population / Nümunə və Əhalinin Anlaşılması

1. **Population: Think of Everyone / Əhali: Hamını Nəzərdə Tutun**
   - The **population** is everyone in a group you're interested in. For example, if you want to know what every student in your school likes for lunch, all the students are your population.
   - **Əhali** sizin maraqlandığınız qrupdakı hər kəsidir. Məsələn, məktəbinizdəki hər bir şagirdin nahar üçün nəyi sevdiyini bilmək istəsəniz, bütün şagirdlər sizin əhalinizdir.

2. **Sample: Just a Few / Nümunə: Yalnız Bir Neçə**
   - A **sample** is just a part of the whole group. If you ask just a few of your classmates about their favorite lunch, those classmates are your sample.
   - **Nümunə** bütün qrupun yalnız bir hissəsidir. Əgər yalnız bir neçə sinif yoldaşınıza sevimli naharlarını soruşsanız, bu sinif yoldaşlarınız sizin nümunənizdir.

3. **Why Samples? / Niyə Nümunələr?**
   - We use samples because it's faster and easier than asking everyone, especially when there are a lot of people in the population.
   - Biz nümunələrdən istifadə edirik, çünki bu, xüsusilə əhali çox olduqda, hər kəsi sorğulamaqdan daha sürətli və asandır.

4. **Choosing Fairly / Ədalətli Seçim**
   - It’s important to pick a sample that includes different types of people to get a fair idea of what everyone might think or like.
   - Hər kəsin nə düşündüyünü və ya nəyi xoşladığını ədalətli bir şəkildə anlamaq üçün müxtəlif növ insanları əhatə edən bir nümunə seçmək vacibdir.

5. **Learning from Samples / Nümunələrdən Öyrənmə**
   - What we learn from the sample can help us guess what the whole population might prefer or think.
   - Nümunədən öyrəndiklərimiz bütün əhalinin nəyi üstün tutacağı haqqında bizə fikir verə bilər.

###**Simple Random Sample**
- Each member of the population has an equal chance of being selected.
- Populyasiyanın hər bir üzvü seçilmək üçün bərabər şansa malikdir.
### **Systematic Sample**
- Members from the population are selected at a regular interval, defined by a step size.
- For example, selecting every 10th person on a list after a random start.
- Populyasiya üzvləri, müəyyən edilmiş addım ölçüsü ilə düzənli intervalda seçilir.
###**Stratified Sample**
- The population is divided into subgroups (strata) that share similar characteristics.
- Populyasiya oxşar xüsusiyyətləri olan alt qruplara (stratalara) bölünür.
###**Cluster Sample**
- The population is divided into separate groups (clusters), usually geographically or by similar characteristics.
- A random sample of these clusters is selected, and all individuals within the chosen clusters are sampled.
- Populyasiya adətən coğrafiya və ya oxşar xüsusiyyətlərə görə ayrı qruplara (klasterlərə) bölünür.



```python
import numpy as np
data = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60, 48, 50, 70, 80, 92, 55, 25, 18, 17, 16, 78, 45, 55, 23, 67, 65, 68, 34, 45,67, 45,33, 88]
def sample_mean(data, sample_size):
    sample = np.random.choice(data, sample_size, replace=False)
    return np.mean(sample)
```

```python
def biased_sample_mean(data, sample_size):
    return np.mean(data[:sample_size])

sample_size = 10
mean1 = biased_sample_mean(data, sample_size)  # This will introduce systematic error
mean2 = sample_mean(data, sample_size)         # This is a proper random sample mean

# Print the results
print(f"Mean of Population: {np.mean(data):.2f}")
print(f"Mean of Sample 1 (Biased): {mean1:.2f}")
print(f"Mean of Sample 2 (Random): {mean2:.2f}")
```

## Law of Large Numbers

```python
import numpy as np
population_size = 10000
population = np.random.randint(1, 101, size=population_size)
print(population_size)
```

```python
print(population[:10])
```

```python
sample_size = 7000
sample1 = np.random.choice(population, sample_size, replace=False)
sample2 = np.random.choice(population, sample_size, replace=False)
population_mean = np.mean(population)
sample1_mean = np.mean(sample1)
sample2_mean = np.mean(sample2)
print(f"Population Mean: {population_mean:.2f}")
print(f"Sample 1 Mean: {sample1_mean:.2f}")
print(f"Sample 2 Mean: {sample2_mean:.2f}")
```

**English:**

As the population grows, increasing the sample size improves the accuracy of the sample mean, but only up to a point. Beyond that, further increases bring diminishing returns and don't significantly improve accuracy.

**Azərbaycan dili:**

Populyasiya artdıqca, nümunə ölçüsünün artırılması nümunə ortalamasının dəqiqliyini artırır, lakin müəyyən bir nöqtəyə qədər. Bundan sonra əlavə artım dəqiqliyi əhəmiyyətli dərəcədə artırmır.

### Common sources of systematic errors in statistics:

- **Selection Bias (Seçim Təhrifləri)**
  - Occurs when the sample is not representative of the population, leading to incorrect conclusions.
  - Nümunə əhalinin nümayəndəsi olmadıqda yaranır və səhv nəticələrə səbəb olur.
  
- **Measurement Bias (Ölçmə Təhrifləri)**
  - Arises from systematic errors in measuring variables, causing inaccurate data collection.
  - Dəyişənləri ölçərkən sistematik səhvlərdən yaranır və qeyri-dəqiq məlumat toplanmasına səbəb olur.
  
- **Instrument Bias (Alət Təhrifləri)**
  - Results from flaws or inconsistencies in the tools or instruments used for data collection.
  - Məlumat toplama üçün istifadə olunan alətlərin qüsurları və ya uyğunsuzluqları nəticəsində yaranır.
  
- **Observer Bias (Müşahidəçi Təhrifləri)**
  - Happens when the researcher's expectations influence the recording or interpretation of data.
  - Tədqiqatçının gözləntiləri məlumatların qeyd edilməsinə və ya şərh edilməsinə təsir etdikdə yaranır.
  
- **Recall Bias (Xatırlama Təhrifləri)**
  - Occurs when participants do not accurately remember past events or experiences.
  - İştirakçıların keçmiş hadisələri və ya təcrübələri düzgün xatırlamaması nəticəsində yaranır.
  
- **Publication Bias (Nəşr Təhrifləri)**
  - The tendency to publish positive results over negative or inconclusive ones, skewing the literature.
  - Mənfi və ya qəti olmayan nəticələrə nisbətən müsbət nəticələrin nəşr olunma meyli, ədəbiyyatı təhrif edir.
  
- **Confounding (Qarışdırıcı Faktörlər)**
  - When an outside variable influences both the independent and dependent variables, distorting the results.
  - Xarici bir dəyişənin həm müstəqil, həm də asılı dəyişənlərə təsir edərək nəticələri təhrif etməsi.
  
- **Sampling Bias (Nümunə Təhrifləri)**
  - Similar to selection bias, it occurs when the sampling method leads to a non-representative sample.
  - Seçim təhriflərinə bənzər şəkildə, nümunə seçmə metodunun nümayəndə olmayan nümunəyə səbəb olması.
  
- **Non-response Bias (Cavabsızlıq Təhrifləri)**
  - Arises when individuals who do not participate differ significantly from those who do.
  - İştirak etməyən şəxslərin iştirak edənlərdən əhəmiyyətli dərəcədə fərqləndiyi halda yaranır.
  
- **Attrition Bias (Müddət Təhrifləri)**
  - Occurs when participants drop out of a study, potentially affecting the results.
  - İştirakçıların araşdırmadan çıxması nəticələrin təsirlənməsinə səbəb ola bilər.
  
- **Reporting Bias (Hesabat Təhrifləri)**
  - When only certain results are reported based on the nature of the findings.
  - Nəticələrin təbiətinə görə yalnız müəyyən nəticələrin hesabat verilməsi.
  
- **Confirmation Bias (Təsdiqləmə Təhrifləri)**
  - The tendency to search for, interpret, and remember information that confirms preconceptions.
  - Əvvəlcədən mövcud olan fikirləri təsdiqləyən məlumatları axtarma, şərh etmə və xatırlama meyli.
  
- **Survivorship Bias (Qalib Qalan Təhriflər)**
  - Focusing on successful cases while ignoring those that failed, leading to misleading conclusions.
  - Uğurlu hallara diqqət yetirmək və uğursuz olanları nəzərə almamaq, yanıltıcı nəticələrə gətirib çıxarır.
  
- **Social Desirability Bias (Sosial Arzu Təhrifləri)**
  - When participants respond in a manner they believe is socially acceptable rather than being truthful.
  - İştirakçıların doğru olmaq əvəzinə sosial olaraq qəbul edilən şəkildə cavab verməsi.
  
- **Calibration Errors (Kalibrləmə Səhvləri)**
  - Inaccuracies that arise from improperly calibrated measurement instruments.
  - Yanlış kalibrlənmiş ölçü alətlərindən yaranan qeyri-dəqiqliklər.
  
- **Processing Errors (Emal Səhvləri)**
  - Mistakes that occur during the data processing phase, such as incorrect data handling.
  - Məlumat emalı mərhələsində, məsələn, səhv məlumat idarəçiliyi kimi səhvlər.
  
- **Data Entry Errors (Məlumat Giriş Səhvləri)**
  - Errors introduced when transferring data from one format to another, such as typos or miskeying.
  - Məlumatı bir formatdan digərinə köçürərkən, məsələn, yazı səhvləri və ya yanlış düymə basmaları nəticəsində yaranan səhvlər.
  
- **Model Specification Errors (Model Təhlükəsi Səhvləri)**
  - Occur when the statistical model does not correctly represent the underlying data structure.
  - Statistik modelin əsas məlumat strukturunu düzgün təmsil etməməsi nəticəsində yaranır.
  
- **Omitted Variable Bias (Saxlanılan Dəyişən Təhrifləri)**
  - Happens when a relevant variable is left out of the model, leading to biased estimates.
  - Əhəmiyyətli bir dəyişənin modeldən çıxarılması nəticəsində meylli təxminlərin yaranması.
