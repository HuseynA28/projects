### Understanding Sample and Population / Nmun v halinin Anlalmas

1. **Population: Think of Everyone / hali: Hamn Nzrd Tutun**
   - The **population** is everyone in a group you're interested in. For example, if you want to know what every student in your school likes for lunch, all the students are your population.
   - **hali** sizin maraqlandnz qrupdak hr ksidir. Msln, mktbinizdki hr bir agirdin nahar n nyi sevdiyini bilmk istsniz, btn agirdlr sizin halinizdir.

2. **Sample: Just a Few / Nmun: Yalnz Bir Ne**
   - A **sample** is just a part of the whole group. If you ask just a few of your classmates about their favorite lunch, those classmates are your sample.
   - **Nmun** btn qrupun yalnz bir hisssidir. gr yalnz bir ne sinif yoldanza sevimli naharlarn sorusanz, bu sinif yoldalarnz sizin nmunnizdir.

3. **Why Samples? / Niy Nmunlr?**
   - We use samples because it's faster and easier than asking everyone, especially when there are a lot of people in the population.
   - Biz nmunlrdn istifad edirik, nki bu, xsusil hali ox olduqda, hr ksi sorulamaqdan daha srtli v asandr.

4. **Choosing Fairly / daltli Seim**
   - Its important to pick a sample that includes different types of people to get a fair idea of what everyone might think or like.
   - Hr ksin n dndyn v ya nyi xoladn daltli bir kild anlamaq n mxtlif nv insanlar hat edn bir nmun semk vacibdir.

5. **Learning from Samples / Nmunlrdn yrnm**
   - What we learn from the sample can help us guess what the whole population might prefer or think.
   - Nmundn yrndiklrimiz btn halinin nyi stn tutaca haqqnda biz fikir ver bilr.

###**Simple Random Sample**
- Each member of the population has an equal chance of being selected.
- Populyasiyann hr bir zv seilmk n brabr ansa malikdir.
### **Systematic Sample**
- Members from the population are selected at a regular interval, defined by a step size.
- For example, selecting every 10th person on a list after a random start.
- Populyasiya zvlri, myyn edilmi addm ls il dznli intervalda seilir.
###**Stratified Sample**
- The population is divided into subgroups (strata) that share similar characteristics.
- Populyasiya oxar xsusiyytlri olan alt qruplara (stratalara) blnr.
###**Cluster Sample**
- The population is divided into separate groups (clusters), usually geographically or by similar characteristics.
- A random sample of these clusters is selected, and all individuals within the chosen clusters are sampled.
- Populyasiya adtn corafiya v ya oxar xsusiyytlr gr ayr qruplara (klasterlr) blnr.



```python
import numpy as np
data = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60, 48, 50, 70, 80, 92, 55, 25, 18, 17, 16, 78, 45, 55, 23, 67, 65, 68, 34, 45,67, 45,33, 88]
def sample_mean(data, sample_size):
    sample = np.random.choice(data, sample_size, replace=False)
    return np.mean(sample)
```

```python
def biased_sample_mean(data, sample_size):
    return np.mean(data[:sample_size])

sample_size = 10
mean1 = biased_sample_mean(data, sample_size)  # This will introduce systematic error
mean2 = sample_mean(data, sample_size)         # This is a proper random sample mean

# Print the results
print(f"Mean of Population: {np.mean(data):.2f}")
print(f"Mean of Sample 1 (Biased): {mean1:.2f}")
print(f"Mean of Sample 2 (Random): {mean2:.2f}")
```

## Law of Large Numbers

```python
import numpy as np
population_size = 10000
population = np.random.randint(1, 101, size=population_size)
print(population_size)
```

```python
print(population[:10])
```

```python
sample_size = 7000
sample1 = np.random.choice(population, sample_size, replace=False)
sample2 = np.random.choice(population, sample_size, replace=False)
population_mean = np.mean(population)
sample1_mean = np.mean(sample1)
sample2_mean = np.mean(sample2)
print(f"Population Mean: {population_mean:.2f}")
print(f"Sample 1 Mean: {sample1_mean:.2f}")
print(f"Sample 2 Mean: {sample2_mean:.2f}")
```

**English:**

As the population grows, increasing the sample size improves the accuracy of the sample mean, but only up to a point. Beyond that, further increases bring diminishing returns and don't significantly improve accuracy.

### Common sources of systematic errors in statistics:

- **Selection Bias (Seim Thriflri)**
  - Occurs when the sample is not representative of the population, leading to incorrect conclusions.
  - Nmun halinin nmayndsi olmadqda yaranr v shv nticlr sbb olur.
  
- **Measurement Bias (lm Thriflri)**
  - Arises from systematic errors in measuring variables, causing inaccurate data collection.
  - Dyinlri lrkn sistematik shvlrdn yaranr v qeyri-dqiq mlumat toplanmasna sbb olur.
  
- **Instrument Bias (Alt Thriflri)**
  - Results from flaws or inconsistencies in the tools or instruments used for data collection.
  - Mlumat toplama n istifad olunan altlrin qsurlar v ya uyunsuzluqlar nticsind yaranr.
  
- **Observer Bias (Mahidi Thriflri)**
  - Happens when the researcher's expectations influence the recording or interpretation of data.
  - Tdqiqatnn gzlntilri mlumatlarn qeyd edilmsin v ya rh edilmsin tsir etdikd yaranr.
  
- **Recall Bias (Xatrlama Thriflri)**
  - Occurs when participants do not accurately remember past events or experiences.
  - tiraklarn kemi hadislri v ya tcrblri dzgn xatrlamamas nticsind yaranr.
  
- **Publication Bias (Nr Thriflri)**
  - The tendency to publish positive results over negative or inconclusive ones, skewing the literature.
  - Mnfi v ya qti olmayan nticlr nisbtn msbt nticlrin nr olunma meyli, dbiyyat thrif edir.
  
- **Confounding (Qardrc Faktrlr)**
  - When an outside variable influences both the independent and dependent variables, distorting the results.
  - Xarici bir dyinin hm mstqil, hm d asl dyinlr tsir edrk nticlri thrif etmsi.
  
- **Sampling Bias (Nmun Thriflri)**
  - Similar to selection bias, it occurs when the sampling method leads to a non-representative sample.
  - Seim thriflrin bnzr kild, nmun sem metodunun nmaynd olmayan nmuny sbb olmas.
  
- **Non-response Bias (Cavabszlq Thriflri)**
  - Arises when individuals who do not participate differ significantly from those who do.
  - tirak etmyn xslrin itirak ednlrdn hmiyytli drcd frqlndiyi halda yaranr.
  
- **Attrition Bias (Mddt Thriflri)**
  - Occurs when participants drop out of a study, potentially affecting the results.
  - tiraklarn aradrmadan xmas nticlrin tsirlnmsin sbb ola bilr.
  
- **Reporting Bias (Hesabat Thriflri)**
  - When only certain results are reported based on the nature of the findings.
  - Nticlrin tbitin gr yalnz myyn nticlrin hesabat verilmsi.
  
- **Confirmation Bias (Tsdiqlm Thriflri)**
  - The tendency to search for, interpret, and remember information that confirms preconceptions.
  - vvlcdn mvcud olan fikirlri tsdiqlyn mlumatlar axtarma, rh etm v xatrlama meyli.
  
- **Survivorship Bias (Qalib Qalan Thriflr)**
  - Focusing on successful cases while ignoring those that failed, leading to misleading conclusions.
  - Uurlu hallara diqqt yetirmk v uursuz olanlar nzr almamaq, yanltc nticlr gtirib xarr.
  
- **Social Desirability Bias (Sosial Arzu Thriflri)**
  - When participants respond in a manner they believe is socially acceptable rather than being truthful.
  - tiraklarn doru olmaq vzin sosial olaraq qbul ediln kild cavab vermsi.
  
- **Calibration Errors (Kalibrlm Shvlri)**
  - Inaccuracies that arise from improperly calibrated measurement instruments.
  - Yanl kalibrlnmi l altlrindn yaranan qeyri-dqiqliklr.
  
- **Processing Errors (Emal Shvlri)**
  - Mistakes that occur during the data processing phase, such as incorrect data handling.
  - Mlumat emal mrhlsind, msln, shv mlumat idariliyi kimi shvlr.
  
- **Data Entry Errors (Mlumat Giri Shvlri)**
  - Errors introduced when transferring data from one format to another, such as typos or miskeying.
  - Mlumat bir formatdan digrin krrkn, msln, yaz shvlri v ya yanl dym basmalar nticsind yaranan shvlr.
  
- **Model Specification Errors (Model Thlksi Shvlri)**
  - Occur when the statistical model does not correctly represent the underlying data structure.
  - Statistik modelin sas mlumat strukturunu dzgn tmsil etmmsi nticsind yaranr.
  
- **Omitted Variable Bias (Saxlanlan Dyin Thriflri)**
  - Happens when a relevant variable is left out of the model, leading to biased estimates.
  - hmiyytli bir dyinin modeldn xarlmas nticsind meylli txminlrin yaranmas.
